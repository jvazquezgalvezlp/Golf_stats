# Bring Back dataset

```python
df = pd.read_csv('data/scraped_df.csv')
```

```python
df.drop(['Unnamed: 0'], axis=1, inplace=True)
```

# Convert string objects into numbers:

```python
df['TOTAL ROUNDS'] = df['TOTAL ROUNDS'].astype(float)

df['TOTAL ADJUSTMENT'] = df['TOTAL ADJUSTMENT'].str.replace(" ","")
df['TOTAL ADJUSTMENT'] = df['TOTAL ADJUSTMENT'].astype(float)

df['TOTAL STROKES'] = df['TOTAL STROKES'].str.replace(",","")
df['TOTAL STROKES'] = df['TOTAL STROKES'].astype(float)

df['TOTAL DRIVES'] = df['TOTAL DRIVES'].astype(float)

df['TOTAL DISTANCE'] = df['TOTAL DISTANCE'].str.replace(",","")
df['TOTAL DISTANCE'] = df['TOTAL DISTANCE'].astype(float)

df['FAIRWAYS HIT'] = df['FAIRWAYS HIT'].str.replace(" ","")
df['FAIRWAYS HIT'] = df['FAIRWAYS HIT'].str.replace(",","")
df['FAIRWAYS HIT'] = df['FAIRWAYS HIT'].astype(float)

df['POSSIBLE FAIRWAYS'] = df['POSSIBLE FAIRWAYS'].str.replace(",","")
df['POSSIBLE FAIRWAYS'] = df['POSSIBLE FAIRWAYS'].astype(float)

df['GREENS HIT'] = df['GREENS HIT'].str.replace(" ","")
df['GREENS HIT'] = df['GREENS HIT'].str.replace(",","")
df['GREENS HIT'] = df['GREENS HIT'].astype(float)

df['# HOLES'] = df['# HOLES'].str.replace(",","")
df['# HOLES'] = df['# HOLES'].astype(float)
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

# Create necessary columns:

```python
df['SCORE AVERAGE'] = (df['TOTAL STROKES']+df['TOTAL ADJUSTMENT'])/df['TOTAL ROUNDS']
df['DRIVING DISTANCE'] = df['TOTAL DISTANCE']/df['TOTAL DRIVES']
df['DRIVING ACCURACY'] = df['FAIRWAYS HIT']/df['POSSIBLE FAIRWAYS']
df['DRIVING USAGE'] = df['TOTAL DRIVES']/df['# HOLES']
df['GREENS IN REGULATION %'] = df['GREENS HIT']/df['# HOLES']
df['SG:PUTTING'] = df['TOTAL SG:PUTTING']/ df['TOTAL ROUNDS']
df['SG: TEE TO GREEN'] = df['SG:OTT']+df['SG:APR']+df['SG:ARG']
```

# Drop unnecessary columns:

# Change NaN values to 0 in # OF WINS, # OF TOP-10S, 1ST, 2ND, 3RD:

```python
df['# OF WINS'].fillna(0, inplace = True)
df['# OF TOP-10S'].fillna(0, inplace = True)
df['1ST'].fillna(0, inplace = True)
df['2ND'].fillna(0, inplace = True)
df['3RD'].fillna(0, inplace = True)
```

```python
df.rename(columns={"TOTAL ADJUSTMENT" : "ADJUSTMENT", "# HOLES" : "HOLES", "TOTAL SG:PUTTING" : "SG_PUTTING", "# OF WINS" : "WINS", "# OF TOP-10S" : "TOP_10S", "SCORE AVERAGE" : "SCORE_AVERAGE", "DRIVING DISTANCE" : "DRIVING_DISTANCE", "DRIVING ACCURACY" : "DRIVING_ACCURACY", "DRIVING USAGE" : "DRIVING_USAGE", "GREENS IN REGULATION %" : "GREENS_IN_REGULATION", "SG:OTT": "SG_OTT", "SG:APR":"SG_APR", "SG:ARG":"SG_ARG", "SG: TEE TO GREEN" : "TOTAL_SG", "RELATIVE/PAR": "RELATIVE_PAR"}, inplace = True)

```
