```python
df_golf_results = pd.read_csv('df_golf_results.csv')
```

```python
df_golf_results.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
df_golf_results.head(30)
```

```python
df_golf_results = df_golf_results.rename(columns = {'0': 'Player_Name'})
```

```python
actual_winners = df_golf_results.Player_Name.head(20).unique()
```

```python
actual_winners
```

```python
most_probable_winners
```

```python
i = 0
for x in actual_winners:
    if x in most_probable_winners:
        i += 1
        print(x)
        print(i)
```
