```python
import numpy as np 
import pandas as pd
import matplotlib as plt
import seaborn as sns
import requests
import json
from bs4 import BeautifulSoup
```

```python
url = 'https://www.pgatour.com/content/pgatour/stats/stat.120.y2019.html'
#url = 'https://www.pgatour.com/stats/stat.120.y2020.html'
```

```python
response = requests.get(url)
```

```python
response.status_code
```

```python
soup = BeautifulSoup(response.content)
```

```python
soup.title.text.strip()
```

```python
player = soup.find_all('td', class_= 'player-name')
```

```python
p = player[0]
```

```python
p.find('a').text.strip()
```

```python
data_names = [p.find('a').text.strip()for p in player]
```

```python
data_names
```

```python
df_golf_results = pd.DataFrame(data_names)
```

```python
df_golf_results.head(30)
```
